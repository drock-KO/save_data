```python
import numpy as np

# matplotlibの読み込み、pltという変数から呼び出せるようにする
import matplotlib.pyplot as plt
# matplotlibによるグラフ描画をjupyter notebook上でも行えるようにする
%matplotlib inline
```

```python
x = np.arange(0, 30)
y = np.arange(60, 90)
plt.xlim(x.min(), x.max())
plt.ylim(y.min(), y.max())
plt.plot(x,y)
```

```python
np.random.shuffle(x)
print(x)
```

```python
plt.scatter(x,y,c="red",marker="o")
```
